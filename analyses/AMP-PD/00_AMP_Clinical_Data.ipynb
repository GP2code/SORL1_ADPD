## Variables used 
- `${ANCESTRY}` = EUR, AFR, AMR, AAC, AJ, MDE, SAS, CAS, EAS, FIN, CAH
- `chr${}:Position:A1:A2`= Chromosom's number, position, reference and alternative alleles
- `${}= 1-7` for 7 variants
- `${Sample id}`= Sample ids of individuals have interest variants

```python
import pandas as pd
import os

sample_ids = [ ] # list of sample IDs

sample_ids = [sid.strip() for sid in sample_ids]

df = pd.read_csv("/${WORK_DIR}/MDS_UPDRS_Part_I.csv")

df['participant_id'] = df['participant_id'].astype(str).str.strip()

output_dir = "individual_sample_tables_I"
os.makedirs(output_dir, exist_ok=True)

for sid in sample_ids:
    sample_df = df[df['participant_id'] == sid]
    if not sample_df.empty:
        output_path = os.path.join(output_dir, f"{sid}.csv")
        sample_df.to_csv(output_path, index=False)
        print(f"Saved table for {sid} with {len(sample_df)} rows.")
    else:
        print(f"No data found for {sid}.")
```
